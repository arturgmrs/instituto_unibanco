## Qual o número de escolas por dependência adminsitrativa? 

```python
import plotly.express as px

# cria um dicionário para associar cada valor numérico a sua categoria
categorias = {
    1: 'Federal',
    2: 'Estadual',
    3: 'Municipal',
    4: 'Privada'
}

# aplica a associação à coluna TP_DEPENDENCIA_ADM_ESCOLA usando o método map()
df_go['categoria'] = df_go['TP_DEPENDENCIA_ADM_ESCOLA'].map(categorias)

# conta os valores na coluna categoria e calcula o número total de escolas
contagem = df_go['categoria'].value_counts()
total_escolas = len(df_go)

# cria uma lista de strings com o número total de escolas e o nome de cada categoria
texto = [f"{contagem[c]} ({(contagem[c]/total_escolas*100):.1f}%) - {c}" for c in contagem.index]

# cria o gráfico de barras interativo usando Plotly e ajusta a largura e altura das barras
fig = px.bar(x=contagem.index, y=contagem.values, 
             text=texto,
             labels={'x': 'Categoria', 
                     'y': 'Quantidade'},
             title='Distribuição escolar do ENEM em Goiás',
             width=700, height=700, template='seaborn')

# atualiza o layout do gráfico para definir o tamanho da fonte do texto da barra e da legenda
fig.update_layout(
    font=dict(size=12)
)

# exibe o gráfico
fig.show()
```

```python
# selecionar linhas com valor 4 na coluna TP_DEPENDENCIA_ADM_ESCOLA
filtro = df_go['TP_DEPENDENCIA_ADM_ESCOLA'] == 4
df_filtrado = df_go[filtro]

# exibir lista em NO_ESCOLA_EDUCACENSO
lista_escolas = df_filtrado['NO_ESCOLA_EDUCACENSO'].tolist()

print(lista_escolas)
```

## Distribuição das escolas em zona rural e urbana

```python
import plotly.express as px

# cria um dicionário para associar cada valor numérico a sua categoria
categorias = {
    1: 'Urbana',
    2: 'Rural',
}

# aplica a associação à coluna TP_DEPENDENCIA_ADM_ESCOLA usando o método map()
df_go['categoria'] = df_go['TP_LOCALIZACAO_ESCOLA'].map(categorias)

# conta os valores na coluna categoria e calcula o número total de escolas
contagem = df_go['categoria'].value_counts()
total_escolas = len(df_go)

# cria uma lista de strings com o número total de escolas e o nome de cada categoria
texto = [f"{contagem[c]} ({(contagem[c]/total_escolas*100):.1f}%) - {c}" for c in contagem.index]

# cria o gráfico de barras interativo usando Plotly e ajusta a largura e altura das barras
fig = px.bar(x=contagem.index, y=contagem.values, 
             text=texto,
             labels={'x': 'Categoria', 
                     'y': 'Quantidade'},
             title='Distribuição escolas por localização do ENEM em Goiás',
             width=700, height=700, template='seaborn')

# atualiza o layout do gráfico para definir o tamanho da fonte do texto da barra e da legenda
fig.update_layout(
    font=dict(size=12)
)

# exibe o gráfico
fig.show()
```

```python
import plotly.graph_objs as go
from plotly.subplots import make_subplots

# Cria os dados para o gráfico de barras
x = ['A', 'B', 'C']
y1 = [10, 20, 30]
bar_trace = go.Bar(x=x, y=y1, name='Gráfico de Barras')

# Cria os dados para o gráfico de dispersão
y2 = [20, 30, 40]
scatter_trace = go.Scatter(x=x, y=y2, mode='markers', name='Gráfico de Dispersão')

# Cria o objeto de figura com os dois subplots
fig = make_subplots(rows=2, cols=1, shared_xaxes=True)

# Adiciona o gráfico de barras ao subplot superior
fig.add_trace(bar_trace, row=1, col=1)

# Adiciona o gráfico de dispersão ao subplot inferior
fig.add_trace(scatter_trace, row=2, col=1)

# Define o título geral do painel
fig.update_layout(title='Painel com Gráfico de Barras e Gráfico de Dispersão')

# Mostra o painel combinado
fig.show()
```
